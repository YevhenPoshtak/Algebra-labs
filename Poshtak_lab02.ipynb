### 3. Підгрупи і гратка підгруп

Для кожної з груп $\mathbb{Z}_{24}$, $Q_8$, $D_5$, $S_4$ виконайте наступні кроки:

1. Створіть групу.
2. Знайдіть усі її підгрупи.   % G.subgroups()
3. Побудуйте гратку підгруп.   % G.subgroup_lattice() може не працювати, запитайте Gemini or ChatGpt
4. Вкажіть групу, яка сама не абелева, але всі її власні підгрупи є абелевими.



```python
# Ваш код тут:
from sage.all import *

def analyze_Z24():
    print("=== Група Z_24 ===")
    G = CyclicPermutationGroup(24)
    
    print(f"Порядок: {G.order()}")
    print(f"Чи абелева? {G.is_abelian()}")
    
    subgroups = G.subgroups()
    print(f"\nКількість підгруп: {len(subgroups)}")
    print("Підгрупи (порядок та генератори):")
    for i, H in enumerate(subgroups):
        print(f"  {i+1}. Порядок {H.order()}: {H.gens()}")
    
    subs = subgroups
    edges = []
    for i, H in enumerate(subs):
        for j, K in enumerate(subs):
            if H != K and H.is_subgroup(K):
                has_intermediate = False
                for L in subs:
                    if L != H and L != K and H.is_subgroup(L) and L.is_subgroup(K):
                        has_intermediate = True
                        break
                if not has_intermediate:
                    edges.append((H, K))
    
    lattice = DiGraph(edges)
    labels = {H: f"{H.order()}" for H in subs}
    
    print("\nГратка підгруп (числа позначають порядок підгрупи):")
    lattice.show(
        vertex_labels=labels,
        vertex_size=500,
        vertex_color='lightblue',
        layout='circular',
        figsize=[6,6],
        title="Гратка підгруп Z_24"
    )

analyze_Z24()
```

```python
# Ваш код тут:
from sage.all import *

def analyze_S4():
    print("=== Група S_4 ===")
    G = SymmetricGroup(4)
    
    print(f"Порядок: {G.order()}")
    print(f"Чи абелева? {G.is_abelian()}")
    
    subgroups = G.subgroups()
    print(f"\nКількість підгруп: {len(subgroups)}")
    print("Перші 10 підгруп (порядок та генератори):")
    for i, H in enumerate(subgroups[:30]):
        print(f"  {i+1}. Порядок {H.order()}: {H.gens()}")
    
    if len(subgroups) > 30:
        print(f"  ... і ще {len(subgroups) - 10} підгруп")
    
    all_proper_abelian = True
    non_abelian_proper = []
    for H in subgroups:
        if H != G and H.order() > 1:
            if not H.is_abelian():
                all_proper_abelian = False
                non_abelian_proper.append(H)
    
    print(f"\nАналіз критерію:")
    print(f"- Група абелева: {G.is_abelian()}")
    print(f"- Всі власні підгрупи абелеві: {all_proper_abelian}")
    
    if not all_proper_abelian:
        print(f"- Знайдено неабелеві власні підгрупи:")
        for H in non_abelian_proper[:3]:  
            print(f"  Порядок {H.order()}: {H.gens()}")
    
    if not G.is_abelian() and all_proper_abelian:
        print("Ця група абелева!")
    else:
        print("Ця група НЕ абелева")
    
    main_subgroups = [H for H in subgroups if H.order() in [1, 2, 3, 4, 6, 8, 12, 24]]
    
    edges = []
    for i, H in enumerate(main_subgroups):
        for j, K in enumerate(main_subgroups):
            if H != K and H.is_subgroup(K):
                has_intermediate = False
                for L in main_subgroups:
                    if L != H and L != K and H.is_subgroup(L) and L.is_subgroup(K):
                        has_intermediate = True
                        break
                if not has_intermediate:
                    edges.append((H, K))
    
    lattice = DiGraph(edges)
    labels = {H: f"ord={H.order()}" for H in main_subgroups}
    
    print("\nСпрощена гратка підгруп (основні порядки):")
    lattice.show(
        vertex_labels=labels,
        vertex_size=800,
        vertex_color='lightyellow',
        layout='spring',
        figsize=[8,8],
        title="Гратка підгруп S_4 (основні)"
    )

analyze_S4()
```

```python
# Ваш код тут:
print("=== Група Z_24 ===")
Z24 = AbelianGroup([24], names='a')
a = Z24.gen(0)
print("Порядок групи:", Z24.order())
print("Чи абелева?", Z24.is_abelian())

gens_Z24 = Z24.gens()
print("Деяка система твірних:", gens_Z24)

min_gens_Z24 = []
for g in Z24:
    if g.order() == Z24.order():
        min_gens_Z24.append([g])
print("Всі мінімальні системи твірних:", min_gens_Z24)

import random
elements = list(Z24)
el1, el2 = random.sample(elements, 2)
H_Z24 = Z24.subgroup([el1, el2])
print(f"Породжена підгрупа елементами {el1}, {el2}:")
print("  Порядок підгрупи:", H_Z24.order())
print("  Генератори підгрупи:", H_Z24.gens())
print()
```

```python
# Ваш код тут:
print("=== Група Q_8 ===")
Q8 = QuaternionGroup()
generators = Q8.gens()
print("Генератори групи:", generators)
print("Порядок групи:", Q8.order())
print("Чи абелева?", Q8.is_abelian())

gens_Q8 = Q8.gens()
print("Деяка система твірних:", gens_Q8)

min_gens_Q8 = []
elements = list(Q8)
for g1 in elements:
    for g2 in elements:
        if g1 != g2 and g1 != Q8.one() and g2 != Q8.one():
            H = Q8.subgroup([g1, g2])
            if H.order() == Q8.order():
                system = sorted([g1, g2], key=lambda x: elements.index(x))
                if system not in min_gens_Q8:
                    min_gens_Q8.append(system)
print("Всі мінімальні системи твірних:", min_gens_Q8)

i = Q8.gens()[0]
j = Q8.gens()[1]
k = i * j
print(f"i = {i}, j = {j}, k = {k}")
print(f"i² = {i**2}")
print(f"j² = {j**2}")
print(f"k² = {k**2}")
print(f"i*j*k = {i*j*k}")

import random
elements = list(Q8)
el1, el2 = random.sample(elements, 2)
H_Q8 = Q8.subgroup([el1, el2])
print(f"Породжена підгрупа елементами {el1}, {el2}:")
print("  Порядок підгрупи:", H_Q8.order())
print("  Генератори підгрупи:", H_Q8.gens())
print()
```
