### 1. Розширення полів

1. Побудуйте розширення $K=\mathbb{Q}(\sqrt[3]{2},i)$ поля $\mathbb{Q}$.
2. Знайдіть степінь розширення $[K:\mathbb{Q}]$ та базис $K$ над $\mathbb{Q}$.
3. Задайте елемент $a=\sqrt[3]{2}+i$ і знайдіть його мінімальний многочлен $m_a(x)$ над полем $\mathbb{Q}$.
4. Побудуйте розширення $L=\mathbb{Q}(\sqrt[3]{2}+i)$ як факторкільце $\mathbb{Q}[x]/(m_a(x))$.
5. Перевірте, чи є поля $K$ та $L$ рівними, порівнявши степені розширення.



```python
x = polygen(QQ)

# 1. Побудова K = Q(∛2, i)
K1.<a> = NumberField(x^3 - 2)  
K.<b> = K1.extension(x^2 + 1)  
print(f"1. K = Q(∛2, i): {K}")

# 2. Степінь і базис
print(f"\n2. [K:Q] = {K.absolute_degree()}")
print("Базис: 1, ∛2, ∛2², i, i∛2, i∛2²")

# 3. Мінімальний многочлен для ∛2 + i
a_elem = a + b
m_a = a_elem.absolute_minpoly()
print(f"\n3. Мінімальний многочлен для ∛2+i:")
print(f"m_a(x) = {m_a}")

# 4. Побудова L = Q(∛2 + i)
L.<c> = NumberField(m_a)
print(f"\n4. L = Q(∛2+i): {L}")

# 5. Порівняння K і L
print(f"\n5. [K:Q] = {K.absolute_degree()}, [L:Q] = {L.absolute_degree()} =>")
print(f"=> K = L: {K.absolute_degree() == L.absolute_degree()}")
```

### 2. Поля розкладу

1. Побудуйте поле розкладу $L$ многочлена $f(x)=x^4+x^2+1$ над полем $\mathbb{Q}$.
2. Знайдіть степінь та базис $L$ над $\mathbb{Q}$.
3. Розкладіть многочлен $f(x)$ на лінійні множники над полем $L$.
4. Побудуйте поле розкладу $K$ многочлена $g(x)=x^5+3x^2+2$ над полем $\mathbb{Z}_3$.
5. Знайдіть степінь та базис $K$ над $\mathbb{Z}_3$.
6. Розкладіть многочлен $g(x)$ на лінійні множники над полем $K$.

```python
# 1. Поле розкладу f(x) = x^4 + x^2 + 1 над Q
Q.<x> = PolynomialRing(QQ)
f = x^4 + x^2 + 1
L.<a> = f.splitting_field()
print("1. Поле розкладу L =", L)

# 2. Степінь та базис L над Q
print("\n2. Степінь та базис L над Q:")
print("   Степінь [L:Q] =", L.degree())
V, phi, _ = L.vector_space()  # Три значення повертає
print("   Базис L над Q:", [phi(v) for v in V.basis()])

# 3. Розклад f(x) на лінійні множники над L
print("\n3. Розклад f(x) на лінійні множники над L:")
print("   f(x) =", f.change_ring(L).factor())

# 4. Поле розкладу g(x) = x^5 + 2 над Z3
print("\n4. Поле розкладу g(x) над Z3:")
Z3 = GF(3)
R.<x> = PolynomialRing(Z3)
g = x^5 + 2
K.<b> = g.splitting_field()
print("   K =", K)
print("   |K| =", K.order())

# 5. Степінь та базис K над Z3
print("\n5. Степінь та базис K над Z3:")
print("   Степінь [K:Z3] =", K.degree())
print("   Базис K над Z3: 1, b, b^2, b^3")

# 6. Розклад g(x) на лінійні множники над K
print("\n6. Розклад g(x) на лінійні множники над K:")
print("   g(x) =", g.change_ring(K).factor())
```

### 3. Скінченні поля

1. Побудуйте поле $F$ з $8$ елементів. Виведіть таблицю множення.
2. Побудуйте факторкільця $K=\mathbb{Z}_2[x]/(x^3+x+1)$ та $L=\mathbb{Z}_2[x]/(x^3+x^2+1)$.
3. Перевірте, що $K$ та $L$ є полями з $8$ елементів.
4. Знайдіть первісні елементи для полів $K$ та $L$.
5. Встановіть ізоморфізм між полями $K$ та $L$.


```python
# 1. Поле з 8 елементів
print("1. Поле GF(8):")
F.<a> = GF(8)
print("Елементи:", list(F))
print()

# 2. Факторкільця 
print("2. Факторкільця:")
R.<x> = PolynomialRing(GF(2))
K.<alpha> = R.quotient(x^3 + x + 1)
L.<beta> = R.quotient(x^3 + x^2 + 1)
print("K = Z2[x]/(x^3+x+1):", list(K))
print("L = Z2[x]/(x^3+x^2+1):", list(L))
print()

#3. Перевірка полів з 8 елементів
print("3 Перевірка полів з 8 елементів:")
print("Чи K поле  8 елементів?", K.is_field() and len(list(K)) == 8, "| Чи L поле з 8 елементів?", L.is_field() and len(list(L)) == 8)
print()

# 4. Первісні елементи
print("4. Первісні елементи:")
for el in K:
    if el != 0:
        pow = el
        order = 1
        while pow != 1:
            pow *= el
            order += 1
        if order == 7:
            print("У K:", el)
            break

for el in L:
    if el != 0:
        pow = el
        order = 1
        while pow != 1:
            pow *= el
            order += 1
        if order == 7:
            print("У L:", el)
            break
print()

# 5. Ізоморфізм
print("5. Ізоморфізм K ≅ L:")
f = x^3 + x + 1
for el in L:
    if el != 0 and f(el) == 0:
        print(f"alpha → {el} (бо ({el})³ + ({el}) + 1 = 0)")
        print(f"Перевірка: alpha² → {el}^2 = {el^2}")
        break
```
