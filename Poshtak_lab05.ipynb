### 1. Абелеві групи




1. Напишіть функцію, яка для заданого $n$ повертає кількість попарно неізоморфних абелевих груп порядку $n$.
2. Знайдіть перші сто значень $n$, для яких існує рівно одна абелева група порядку $n$, і порівняйте з розкладом $n$ у добуток простих.
3. Напишіть гіпотезу про те, для яких $n$ існує рівно одна абелева група порядку $n$.
4. Спробуйте знайти всі $n$, для яких існує рівно $n$ абелевих груп порядку $n$.

```python
def count_abelian_groups(n):
    if n == 1:
        return 1
    total = 1
    for p, exp in factor(n):
        total *= number_of_partitions(exp)
    return total

print("2. Перші 100 n з однією абелевою групою:")
single_groups = []
n = 1
while len(single_groups) < 100:
    if count_abelian_groups(n) == 1:
        single_groups.append(n)
    n += 1
print(single_groups)

print("\nПункт 3: Гіпотеза")
print("Єдина абелева група порядку n існує ⇔ n в розкладі всі степені простих рівні 1")

print("\nПункт 4: n з n абелевими групами (n ≤ 100000):")
special_n = []
for n in range(1, 100001):
    if count_abelian_groups(n) == n:
        special_n.append(n)
print(special_n)
```

### 2. Дільники нуля в кільцях

1. Задайте кільця $\mathbb{Z}_{30}$, $\mathbb{Z}_{12}\times\mathbb{Z}_{15}$, $M_3(\mathbb{Z}_3)$, $\mathbb{Z}_6[x]$.
2. Знайдіть кількість дільники нуля в цих кільцях.
3. Знайдіть мультиплікативну групу та її порядок. 



```python
from functools import lru_cache

Z30 = Integers(30)
zero_divisors_Z30 = [a for a in Z30 if a != 0 and any(a * b == 0 and b != 0 for b in Z30)]
units_Z30 = [a for a in Z30 if gcd(a, 30) == 1]
print(f"1. Z30: дільників нуля - {len(zero_divisors_Z30)}, порядок групи - {len(units_Z30)}")

Z12 = Integers(12)
Z15 = Integers(15)
zero_divisors_product = [(a,b) for a in Z12 for b in Z15 if (a,b) != (0,0) and 
                         any(a*c == 0 and b*d == 0 and (c,d) != (0,0) for c in Z12 for d in Z15)]
units_product = [(a,b) for a in Z12 for b in Z15 if gcd(a,12) == 1 and gcd(b,15) == 1]
print(f"2. Z12×Z15: дільників нуля - {len(zero_divisors_product)}, порядок групи - {len(units_product)}")

Z3 = Integers(3)
M3Z3 = MatrixSpace(Z3, 3)

@lru_cache(maxsize=None)
def is_zero_divisor(matrix_tuple):
    M = M3Z3(matrix_tuple)
    if M == 0:
        return False
    for N in [M3Z3([1,0,0,0,0,0,0,0,0]), M3Z3([0,1,0,0,0,0,0,0,0]), M3Z3([0,0,1,0,0,0,0,0,0])]:
        if M * N == 0 and N != 0:
            return True
    return M.determinant() == 0

matrices_tuples = [tuple(M.list()) for M in M3Z3]
zero_divisors_count = sum(1 for mat_tuple in matrices_tuples if is_zero_divisor(mat_tuple))

GL3 = GL(3, Z3)
print(f"3. M3(Z3): дільників нуля - {zero_divisors_count}, порядок групи - {GL3.order()}")

Z6 = Integers(6)
units_count = len([a for a in Z6 if gcd(a, 6) == 1])
print(f"4. Z6[x]: дільників нуля - ∞, порядок групи - {units_count}")
```

### 3. Корінь з одиниці за модулем $p$

1. Знайдіть перші сто простих чисел $p$, для яких факторкільце $\mathbb{Z}_p[x]/(x^2+1)$ є полем.  $\qquad$ % PolynomialRing(GF(p))
2. Знайдіть остачі при ділення цих простих чисел на $m$ для деяких значень $m=2,3,\ldots,10$.        
3. Спробуйте знайти закономірність і сформулювати гіпотезу про такі прості числа.


```python
primes = []
p = 3  
while len(primes) < 100:
    if is_prime(p) and legendre_symbol(-1, p) == -1:
        primes.append(p)
    p += 1

print("1. Перші 100 простих p, де Zp[x]/(x^2+1) - поле:")
print(primes)

print("\n2. Остачі за модулями:")
for m in range(2, 11):
    remainders = [p % m for p in primes]
    unique = sorted(set(remainders))
    print(f"mod {m}: {unique}")

print(f"\n3. Гіпоеза: підходять yсі p ≡ 3 (mod 4): {all(p % 4 == 3 for p in primes)}")
```

### 4. Ідеали та факторкільця кільця многочленів

1. Задайте кільце многочленів $R=\mathbb{Z}_{7}[x]$. $\qquad$  %  R.\<x\>=Integers(7)[]
3. Задайте ідеали $I_1=(2x^3+5x+1)$, $I_2=(x^4+3x^3+2)$, $I_3=(x^5+x^3+x+1)$.  $\qquad$    %  I = R.ideal(f(x))
4. Побудуйте факторкільця $R/I$ і знайдіть його порядок. $\qquad$  %  R.quotient(І)
5. Перевірте, для яких ідеалів факторкільце $R/I$ є полем.
6. Перевірте, чи має елемент $(x^5+3x^2+2)+I$ обернений у факторкільці $R/I$ та знайдіть обернений.

```python
R.<x> = PolynomialRing(GF(7))
print("1. Кільце многочленів R = Z7[x]")
print("R =", R)
print()

I1 = R.ideal(2*x^3 + 5*x + 1)
I2 = R.ideal(x^4 + 3*x^3 + 2)
I3 = R.ideal(x^5 + x^3 + x + 1)

print("2. Ідеали:")
print("I1 =", I1)
print("I2 =", I2)
print("I3 =", I3)
print()

print("3. Факторкільця та їх порядки:")

Q1 = R.quotient(I1)
print(f"R/I1: порядок = {Q1.order()}")
print(f"   Елементи: {list(Q1)[:10]}{'...' if Q1.order() > 10 else ''}")

Q2 = R.quotient(I2)
print(f"R/I2: порядок = {Q2.order()}")
print(f"   Елементи: {list(Q2)[:10]}{'...' if Q2.order() > 10 else ''}")

Q3 = R.quotient(I3)
print(f"R/I3: порядок = {Q3.order()}")
print(f"   Елементи: {list(Q3)[:10]}{'...' if Q3.order() > 10 else ''}")
print()

print("4. Перевірка, чи є факторкільце полем:")

def is_field(ideal, quotient_ring, name):
    poly = ideal.gens()[0]
    print(f"R/{name}:")
    print(f"   Поліном: {poly}")
    print(f"   Степінь: {poly.degree()}")
    print(f"   Незвідний: {poly.is_irreducible()}")
    print(f"   Факторкільце є полем: {quotient_ring.is_field()}")
    return quotient_ring.is_field()

result1 = is_field(I1, Q1, "I1")
result2 = is_field(I2, Q2, "I2") 
result3 = is_field(I3, Q3, "I3")
print()

print("5. Перевірка оберненості елемента (x^5 + 3x^2 + 2) + I:")

element = x^5 + 3*x^2 + 2
print(f"Елемент для перевірки: {element}")

def check_inverse(ideal, quotient_ring, name):
    try:
        elem_quot = quotient_ring(element)
        if elem_quot == 0:
            print(f"R/{name}: Елемент дорівнює 0 у факторкільці - необернений")
            return None
        else:
            inverse_elem = elem_quot^(-1)
            print(f"R/{name}: Елемент ОБЕРНЕНИЙ")
            print(f"   Обернений елемент: {inverse_elem}")
            check = elem_quot * inverse_elem
            print(f"   Перевірка: {elem_quot} * {inverse_elem} = {check}")
            return inverse_elem
    except (ZeroDivisionError, ArithmeticError) as e:
        print(f"R/{name}: Елемент НЕ ОБЕРНЕНИЙ")
        print(f"   Помилка: {e}")
        return None
    except Exception as e:
        print(f"R/{name}: Помилка при обчисленні: {e}")
        return None

print("\nПеревірка для кожного ідеалу:")
inv1 = check_inverse(I1, Q1, "I1")
inv2 = check_inverse(I2, Q2, "I2")
inv3 = check_inverse(I3, Q3, "I3")
```
